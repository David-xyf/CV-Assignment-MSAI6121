```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
# get the image
def getImage(path):
    return Image.open(path)


# show the image
def showImage(img, isGray=False):
    plt.axis("off")
    if isGray == True:
        plt.imshow(img, cmap="gray")
    else:
        plt.imshow(img)
    plt.show()
```

```python
# histogram equalization algorithm for basic pictures
def histEqAlgo(imgArr):
    hist, bins = np.histogram(imgArr, 255)
    cdf = np.cumsum(hist)
    cdf = 255 * (cdf / cdf[-1])  # mapping
    res = np.interp(imgArr.flatten(), bins[:-1], cdf)
    res = res.reshape(imgArr.shape)
    return res, hist
```

```python
img = getImage("sample01.jpg")
gray_image = img.convert('L')
gray_arr = np.array(gray_image)
res, hist = histEqAlgo(gray_arr)
showImage(Image.fromarray(res))
```

```python
# histogram equalization algorithm for Colorful pictures
def RGB_histEqAlgo(imgArr):
    imgArr = np.array(imgArr)
    imgArr2 = imgArr.flatten()
    hist, bins = np.histogram(imgArr2, 255)
    cdf = np.cumsum(hist)
    cdf = 255 * (cdf / cdf[-1])  # mapping
    res = np.interp(imgArr2, bins[:-1], cdf)
    res = res.reshape(imgArr.shape)
    return Image.fromarray(res.astype("uint8"), mode="RGB")
```

```python
img2 = RGB_histEqAlgo(img)
showImage(img2)
```

```python
# histogram equalization algorithm every channel for Colorful pictures
def RGB_histEqAlgo2(imgArr):
    imgArr = np.array(imgArr)
    r_arr = imgArr[..., 0]
    g_arr = imgArr[..., 1]
    b_arr = imgArr[..., 2]

    r_res, r_hist = histEqAlgo(r_arr)
    g_res, g_hist = histEqAlgo(g_arr)
    b_res, b_hist = histEqAlgo(b_arr)

    new_imarr = np.zeros(imgArr.shape, dtype='uint8')
    new_imarr[..., 0] = r_res
    new_imarr[..., 1] = g_res
    new_imarr[..., 2] = b_res

    return Image.fromarray(new_imarr, mode='RGB')
```

```python
img3 = RGB_histEqAlgo2(img)
showImage(img3)
```

```python
def rgb_histeq3(im):
    imarr = np.array(im)
    r_arr = imarr[..., 0]
    g_arr = imarr[..., 1]
    b_arr = imarr[..., 2]

    #取三个通道的平均值
    imarr2 = np.average(imarr, axis=2)
    hist, bins = np.histogram(imarr2, 255)
    cdf = np.cumsum(hist)
    cdf = 255 * (cdf / cdf[-1])

    r_res = np.interp(r_arr, bins[:-1], cdf)
    g_res = np.interp(g_arr, bins[:-1], cdf)
    b_res = np.interp(b_arr, bins[:-1], cdf)

    new_imarr = np.zeros(imarr.shape, dtype="uint8")
    new_imarr[..., 0] = r_res
    new_imarr[..., 1] = g_res
    new_imarr[..., 2] = b_res

    return Image.fromarray(new_imarr, mode='RGB')


img = getImage("sample01.jpg")
img4 = rgb_histeq3(img)
showImage(img4)
```
