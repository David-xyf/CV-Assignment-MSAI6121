```python
from PIL import Image
import numpy as np
import matplotlib.pyplot as plt
import os
%matplotlib inline
plt.figure(figsize=(10, 10))
for i in range(1, 9):
    plt.subplot(8, 1, i)
    Path = 'images/sample0' + str(i) + '.jpg'
    if Path.split('/')[1] not in os.listdir('images/'):
        Path = 'images/sample0' + str(i) + '.jpeg'

    src = Image.open(Path)
    r, g, b = src.split()
    ar = np.array(r).flatten()
    plt.hist(ar, bins=256, facecolor='r', edgecolor='r')
    ag = np.array(g).flatten()
    plt.hist(ag, bins=256, facecolor='g', edgecolor='g')
    ab = np.array(b).flatten()
    plt.hist(ab, bins=256, facecolor='b', edgecolor='b')

plt.show()
```

```python
plt.figure(figsize=(10, 40))
for i in range(1, 9):
    plt.subplot(8, 1, i)
    Path = 'images/out/res' + str(i) + '.jpg'

    src = Image.open(Path)
    r, g, b = src.split()
    ar = np.array(r).flatten()
    plt.hist(ar, bins=256, facecolor='r', edgecolor='r')
    ag = np.array(g).flatten()
    plt.hist(ag, bins=256, facecolor='g', edgecolor='g')
    ab = np.array(b).flatten()
    plt.hist(ab, bins=256, facecolor='b', edgecolor='b')

plt.show()
```

```python
plt.figure(figsize=(10, 10))
for i in range(1, 9):
    Path = 'images/images_gray/gray_res' + str(i) + '.jpg'
    plt.subplot(8,1,i)
    src = Image.open(Path)
    arr=(np.array(src)).flatten()
    plt.hist(arr, bins=256, facecolor='green', alpha=0.75) 
plt.show()
```
