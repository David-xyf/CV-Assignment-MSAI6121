```python
clahe = cv2.createCLAHE(clipLimit =2.0, tileGridSize=(8,8))
cl_img = clahe.apply(img)
```

```python
plt.hist(cl_img.flat, bins=100, range=(100, 255))
```

```python
Image.fromarray(cl_img)
```

```python
ret,thresh1 = cv2.threshold(cl_img,190,150,cv2.THRESH_BINARY)
ret,thresh2 = cv2.threshold(cl_img,190,255,cv2.THRESH_BINARY_INV)
```

```python
import cv2 as cv 
import numpy as np
 
test=cv.imread("sample01.jpg")
#
B,G,R = cv.split(test) 
clahe = cv.createCLAHE(clipLimit=2.0, tileGridSize=(8, 8))
clahe_B = clahe.apply(B)
clahe_G = clahe.apply(G)
clahe_R = clahe.apply(R)
clahe_test = cv.merge((clahe_B,clahe_G,clahe_R))
cv.imshow('clahe_test',clahe_test)
cv.imshow("test",test)
```

```python
cv.waitKey(0)
cv.destroyAllWindows()
```
